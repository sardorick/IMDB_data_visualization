```python
df['directors'].value_counts().nlargest(10).plot(kind='bar', title='Director with the highest number of films', figsize=(10,10))
plt.xlabel('Name of directors')
plt.ylabel('Number of movies')
plt.savefig('Director with the highest number of films.png')
```

```python
actors  = df.stars.str.split(',')

actors_name = []
for i in actors:
    for j in i:
        actors_name.append(j.replace("'"," ").strip("[ ]"))

df_stars = pd.DataFrame(actors_name)
df_stars.value_counts().nlargest(10).plot(kind='bar', title='Which actor has played the most in Top-100?', figsize=(10,10))
plt.xlabel('Name of actors')
plt.ylabel('Number of movies')
plt.savefig('Actor with the highest number of films.png')
```
